##Lake Model##

(1)

Let $\beta = $ fraction of workers going from New York to Chicago.

Let $\gamma= $ fraction of workers going from Chicago to New York.

Then the laws of motion for $N_{t+1}$, $C_{t+1}$, and $P_{t+1}$ are:

$N_{t+1} = (1-\beta)N_t + \gamma C_t$

$C_{t+1} = (1-\gamma)C_t + \beta N_t$

$P_{t+1} = P_t$ since there are no births or deaths in this particular Lake Model.

(2)

$\begin{bmatrix}
1-\beta & \gamma \\
\beta & 1-\gamma
\end{bmatrix}$

(3)

Because there are no births nor are there deaths, then the growth rate, $g$ = 0. 
Thus normally in this case, $\hat{A} = \frac{1}{1+g}\times A$ but since $g=0$ then $\hat{A}=A$

(4)

To find the steady state, we must find the eigenvector [n,c] for A given the $\lambda = 1$ to satisfy the equation: 

$\hat{A}\left[ \begin{array}{c} n \\ c \end{array} \right] = \lambda \left[ \begin{array}{c} n \\ c \end{array} \right]$ where $\lambda = 1$. 

When $\lambda = 1$ we have the following equation for the nullspace of $\hat{A}$. 

$\begin{bmatrix} -\beta & \gamma \\ \beta & -\gamma \end{bmatrix} \times 
\left[ \begin{array}{c} n \\ c \end{array} \right] = 
\left[ \begin{array}{c} 0 \\ 0 \end{array} \right]$


$\begin{bmatrix} -\beta & \gamma \\ 0 & 0 \end{bmatrix} \times 
\left[ \begin{array}{c} n \\ c \end{array} \right] = 
\left[ \begin{array}{c} 0 \\ 0 \end{array} \right]$


$\begin{bmatrix} 1 & \frac{-\gamma}{\beta} \\ 0 & 0 \end{bmatrix} \times 
\left[ \begin{array}{c} n \\ c \end{array} \right] = 
\left[ \begin{array}{c} 0 \\ 0 \end{array} \right]$


Therefore, $n-\frac{\gamma}{\beta}c = 0$, which implies that the corresponding eigenvector to 
$\lambda = 1$ is c$\left[ \begin{array}{c} \frac{\gamma}{\beta} \\ 1 \end{array} \right]$ where $c \in R$ (reals). Thus given a certain $\gamma$ and $\beta$ the eigenvector should be adjusted accordingly so that n+c = 1.

(5)

The probability that a worker is in Chicago at time $t$ and ends up in New York at time $t+3$ is determined by the transition matrix:

P = $\begin{bmatrix} 1-\beta & \gamma \\ \beta & 1-\gamma \end{bmatrix}$, where $\beta$ represents the transition probability from New York to Chicago, and $\gamma$ represents the transition probability from Chicago to New York.

The four paths by which a worker can start in Chicago and end up in New York from time $t$ to time $t+3$ are as follows:

(1) $C\rightarrow C\rightarrow C \rightarrow N$, which corresponds to the probability $(1-\gamma)^2\gamma$

(2) $C\rightarrow C\rightarrow N \rightarrow N$, which corresponds to the probability $(1-\gamma)(\gamma)(1-\beta)$

(3) $C \rightarrow N \rightarrow C \rightarrow N$, which corresponds to the probability $(\gamma)(\beta)(\gamma)$

(4) $C \rightarrow N \rightarrow N \rightarrow N$, which corresponds to the probability $(\gamma)(1-\beta)^2$

```python
#Simulation

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#Set up Parameters
N = 1000
n0 = 0.5
c0 = 0.5
beta = 0.3
gamma = 0.15

#Set up Transition Matrix
A = np.array(([1-beta,gamma],[beta,1-gamma]))

#Set up Simulation
x_rate = np.zeros((N,2)) #where x is the state variable containing entries, x0 = [n0,c0]
x_rate[0,:] = np.array(([n0,c0]))

#Generate Data
for i in range(1,N):
    x_rate[i,:] = np.dot(A,x_rate[i-1,:])

#Plot
fig, ax = plt.subplots(figsize = (10,5))
ax.plot(x_rate,alpha = 0.8)
plt.show()
```

In the graph, the fraction of workers in New York and Chicago are $.\bar{3}$ and $.\bar{6}$ respectively. 

Thus according to the calculation of the steady state in question 4, we have:

$\left[ \begin{array}{c} n \\ c \end{array} \right] = 
\left[ \begin{array}{c} \frac{\gamma}{\beta} \\ 1 \end{array} \right] = 
\left[ \begin{array}{c} .5 \\ 1 \end{array} \right]$, so we must scale this vector down such that the components add to 1.

Notice that the ratio between the magnitude of the components is 2, thus we could represent this vector more generally as $\left[ \begin{array}{c} x \\ 2x \end{array} \right]$.

Thus if $x+2x = 1$, then $x = \frac{1}{3}$. Therefore, $n = \frac{1}{3}$ and $c=\frac{2}{3}$.

##Markov Chain##

(1)

A matrix is stochastic if the rows sum to 1 (corresponding to the fact that probability densities must sum to 1), and if the row elements are all positive (because you cannot have a negative probability). 

In this case, $P_1$ is not stochastic because the second row does not sum to 1, and neither is $P_2$ because the second row has a negative value.

(2)

*"assuming if one state is periodic then the whole matrix is periodic"

$P_1$ is irreducible because all states can be reached, but is not aperiodic as the periods of the 1st and 2nd state is equal to 2.

$P_2$ is irreducible and aperiodic because all states can be reached and the period for each of the states is equal to one.

$P_3$ is not irreducible because the 3rd state cannot be reached from either of the first two states, and even if the system starts in the 3rd state, it will eventually leave the 3rd state and be unable to return. $P_3$ is also aperiodic because all states have a period equal to one.

$P_4$ is irreducible and aperiodic because all states can be reached and the period for each of the states is equal to one.

```python
#Simulation

#Parameters
#N = Number of States
#T = Number of Periods
#P* = Stochastic Matrices

N = 3
T = 100
P1 = np.array(([0,0,1],[0,0,1],[1/3,1/3,1/3]))
P2 = np.array(([0,1,0],[1/2,0,1/2],[1/4,1/4,1/2]))
P3 = np.array(([0,1,0],[1/2,1/2,0],[1/4,1/4,1/2]))
P4 = np.array(([0,1/2,1/2],[1/2,1/2,0],[1/4,1/4,1/2]))

Px = np.array(([P1,P2,P3,P4]))

#Plot Setup
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
ax1 = axes[0,0]
ax2 = axes[0,1]
ax3 = axes[1,0]
ax4 = axes[1,1]

axx = np.array(([ax1,ax2,ax3,ax4]))

#Set up Simulation
X = np.zeros((T),dtype = np.int)
X[0]=0

Shock = np.random.uniform(0,1,T)

#Simulation
for k in range(0,4): #Iterating through each Markov Chain
    P = Px[k]
    ax = axx[k]
    for i in range(1,T): #Iterating through each period
        X[i]=0
        for j in range(1,N):
            if Shock[i] > sum(P[X[i-1],0:j]) \
            and Shock[i] <= sum(P[X[i-1],0:j+1]):
                X[i] = j
    ax.plot(X,'bo--', lw=2, alpha=0.8, label='Path')
    ax.set_ylabel("P%s" %(k+1), fontsize = 16)

plt.show()
#REMINDER TO FIX THIS BY ADDING 3 MORE SUBPLOTS 
#FOR THE OTHER MARKOV CHAINS
```
